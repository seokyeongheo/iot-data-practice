### functionalize, adapting to whole features

````python
# functionalize

def make_date(word):
    date = re.findall('\d\W\d\W\d\d\d\d', word)
    date = ''.join(date)
    return date
    
def make_string(word):
    string = re.findall('[a-zA-Z]', word)
    string = ''.join(string)
    return string

data_raw['date'] = data_raw['Unnamed: 0'].apply(lambda x : make_date(x))
data_raw['main'] = data_raw['Unnamed: 0'].apply(lambda x : make_string(x))
````

````python
# change index(number) to date
data_raw.index = data_raw.date

# delete 'Unnamed: 0', 'date'
data_revise = data_raw.drop(['Unnamed: 0', 'date'], axis=1)

# fill Nan Value as 'nan'
data_revise.fillna('nan', inplace=True)

# erase 'blank in component'
def delete_blank(word):
    return re.sub(r'\s', '', word)

columns = data_revise.columns

for column in columns:
    data_revise[column] = data_revise[column].apply(lambda x : delete_blank(x))
    
np.unique(data_revise.values), len(np.unique(data_revise.values))
````

````python
from efficient_apriori import apriori as apriori
from apyori import apriori as apriori_pre
import pandas as pd

data_revise = pd.read_pickle('dataset_v6(nan).pkl')
data_revise.shape, data_revise.values[0][0]
````

````python
# recodes (LIL: list of lists 만들기). DataFrame을 apriori 패키지를 활용하여, 연관분석을 하기 위해서는 LIL로 바꿔줘야 한다.
records = []  
for i in range(0, 1499):  
    records.append(tuple([str(data_revise.values[i,j]) for j in range(0, 35)]))

# 학습 : min_support 는 0.1, min_confidence 도 0.1 로 . 
itemsets, rules = apriori(records, min_support=0.1, min_confidence=0.1)
rules_rhs = filter(lambda rule: len(rule.lhs) == 1 and len(rule.rhs) == 1, rules)

result = sorted(rules_rhs, key=lambda rule: rule.lift, reverse=True) # rule.lift - 향상도 - 기준 내림차순 정렬한다.
for i, rule in enumerate(result, start=1):
    print(rule)
    if i == 11: # 상위 10개만 볼 수 있도록.
        break
````
